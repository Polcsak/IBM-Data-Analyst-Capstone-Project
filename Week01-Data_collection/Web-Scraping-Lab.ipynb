## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Download the webpage at the url


```python
#your code goes here

data  = requests.get(url).text
```

Create a soup object


```python
#your code goes here

soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here

table = soup.find('table') # in html table is represented by the tag <table>
```

```python
#Get all rows from the table
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    color_name = cols[2].getText() # store the value in column 3 as color_name
    color_code = cols[3].getText() # store the value in column 4 as color_code
    print("{}--->{}".format(color_name,color_code))
```

Save the scrapped data into a file named *popular-languages.csv*


```python
import pandas as pd

# Seznam pro ukládání dat
data = []

# Get all rows from the table
for row in table.find_all('tr'):  # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td')  # in html a column is represented by the tag <td>
    if len(cols) >= 5:  # Check if there are enough columns
        language_name = cols[1].getText()  # store the value in column 3 as language_name
        annual_average_salary = cols[3].getText()  # store the value in column 5 as annual_average_salary
        # Append data to the list
        data.append([language_name, annual_average_salary])

# Vytvoření DataFrame z dat
df = pd.DataFrame(data,columns=data[0])

# Uložení DataFrame do CSV souboru
df.to_csv('scraped_data.csv', index=False)

print("Data have been saved to scraped_data.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Odeslat požadavek na stránku
url = 'https://zvierata.bazos.sk/'
response = requests.get(url)

# Kontrola, zda byl požadavek úspěšný
if response.status_code == 200:
    # Analyzovat HTML obsah pomocí BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Najít všechny inzeráty
    inzeraty = soup.find_all('div', class_='inzeraty')

    # Seznam pro ukládání dat
    data = []

    # Procházet inzeráty a získat informace
    for inzerat in inzeraty:
        nazev_element = inzerat.find('h2').find('a')
        nazev = nazev_element.text.strip()
        link = nazev_element['href']
        
        cena_element = inzerat.find('div', class_='popis')
        cena = "Cena není uvedena"
        if cena_element:
            # Získejte text popisu
            popis_text = cena_element.text.strip()
            # Pokud se v popisu nachází cena, získejte ji
            if 'cena' in popis_text.lower():
                cena = popis_text.split('cena', 1)[-1].strip().split('\n', 1)[0]
        
        adresa_element = inzerat.find('span', class_='lokalita')
        adresa = adresa_element.text.strip() if adresa_element else "Umístění není uvedeno"
        
        # Přidat data do seznamu
        data.append([nazev, cena, adresa, link])

    # Uložit data do CSV souboru
    with open('inzeraty.csv', 'w', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        # Nastavit hlavičky sloupců
        csvwriter.writerow(['Název', 'Cena', 'Umístění', 'Odkaz'])
        # Zapsat data
        csvwriter.writerows(data)

    print("Data byla uložena do inzeraty.csv")
else:
    print("Chyba při načítání stránky:", response.status_code)
```
